```python
import numpy as np
import matplotlib.pyplot as plt
```

# Սահմանել ֆունկցիան

```python
def func_1(x, y):
    return (x - 2)**2 + (y - 2)**2 + 2
```

```python
def func_2(x, y):
    return np.cos(x)*np.sin(y)
```

```python
def func_3(x, y):
    return x**4 - x**2 + y**2
```

# Սահմանել գրադիենտի ֆուկցիան

```python
def grad_1(x, y):
    return 2*(x - 2), 2*(y - 2)
```

```python
def grad_2(x, y):
    return -np.sin(x)*np.sin(y), np.cos(x)*np.cos(y)
```

```python
def grad_3(x, y):
    return 4*x**3 - 2*x, 2*y
```

# Gradient descent-ի օգնությամբ գտնել էքստրեմումները

Առաջին և երկրորդ ֆունկցիաների համար գտնել մինիմում, երրորդ-ի համար՝ մաքսիմում կետը:

### Քայլ 1: ինչ-որ կետում հաշվել գրադիենտի արժեքը, օրինակ x=5, y=5 կետում

```python
grad_2(5, 5)
```

### Քայլ 2: Սահմանել ինչ-որ learning rate, օրինակ alpha = 0.1 ու կատարել համապատասխան update-ը (շարժվել համապատասխան ուղղությամբ)

```python
func_2(5, 5)
```

```python
func_2(4.4, 4.4)
```

### Քայլ 3: Կատարել Քայլ 2-ը այնքան անգամ, մինչև ֆունկցիայի արժեքը այլևս չի փոփոխվում (չի նվազում կամ աճում): Դրա համար պետք է համեմատել ինչ-որ քայլում ֆունկցիայի արժեքը նախորդ քայլի արժեքի հետ ու եթե տարբերությունը շատ փոքր է (ենթադրենք՝ 0.01-ից փոքր է), ապա դադարեցնել իտերացիաները:

<font color='green'>Հուշում`</font> Դա պետք է անել մի հատ for loop-ի մեջ: Ընդամենը հարկավոր է ստուգել պայման և դադարեցնել ալգորիթմը, երբ այդ պայմանը բավարարվում է: Կարող եք օգտվել հետևյալ էջից՝ [link](https://www.programiz.com/python-programming/break-continue):

```python
x, y = 5, 5
itercount = 1000
alfa = 0.1

for i in range(itercount):
    results = func_2(x, y)
    gradients = grad_2(x, y)
    print ('a' + str(i), results, x, y)
    x, y = x - alfa*gradients[0], y - alfa*gradients[1]
    result_1 = func_2(x, y)
    if abs(results - result_1) < 0.01:
        break
```
